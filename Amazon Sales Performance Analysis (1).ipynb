```python
# Columns to analyze for outliers
columns_to_analyze = ['Total Revenue', 'Total Cost', 'Total Profit']

# Plotting boxplots for each column to analyse outliers
for col in columns_to_analyze:
    plt.figure(figsize=(4, 2))
    sns.boxplot(data[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```

```python
# Columns to analyze for outliers
columns_to_analyze = ['Total Revenue', 'Total Cost', 'Total Profit']
outlier_records = pd.DataFrame()

# Loop through each column
for col in columns_to_analyze:
    Q1 = np.percentile(data[col], 25)
    Q3 = np.percentile(data[col], 75)
    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    upp = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = data[(data[col] < low) | (data[col] > upp)]
    
    # Append outliers to the outlier_records dataframe
    outlier_records = pd.concat([outlier_records, outliers], ignore_index=False)

# Remove outliers from the original data
Clean_data = data[~data.index.isin(outlier_records.index)]

Clean_data
```

Observation:Seven outliers were removed from the dataset, improving data quality by eliminating extreme values that could skew the analysis.
